```python
for s_name in xls.sheet_names[0:9]:               #運用迴圈使111年-103年資料自動整理呈現
   vars()["df_"+s_name] = data.get(s_name)
   vars()["df_"+s_name+"_1"] = vars()["df_"+s_name].iloc[[0],[0,4,5,6,7,8,9,10]]
   vars()["df_"+s_name+"_1"] = vars()["df_"+s_name+"_1"].rename(columns={'Unnamed: 0':"年份"})
   vars()["df_"+s_name+"_1"].loc[0,"年份"] = str(s_name)+"年"
   vars()["df_"+s_name+"_1"] = vars()["df_"+s_name+"_1"].set_index("年份")
   vars()["df_"+s_name+"_2"] = vars()["df_"+s_name+"_1"].T
```
